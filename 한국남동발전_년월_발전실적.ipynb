```python
import pandas as pd


# 엑셀 파일 경로
file_path = '한국남동발전_발전실적.xls'

# 엑셀 파일 읽기
df = pd.read_excel(file_path)

# 데이터 프레임 출력
print(df)
data = pd.read_csv('한국남동발전_발전실적.csv', encoding='cp949')
```

```python
df[df['사업소'] == '경상대태양광']
```

```python
df['일자'] = df['일자'] - 200000
```

```python
use_df = df[df['사업소'] == '경상대태양광']
use_df
```

```python
use_df['년'] = use_df['일자']//100
use_df['월'] = use_df['일자']%100
```

```python
use_df
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
rc('font', family='Malgun Gothic')  # 윈도우 사용 시

plt.figure(figsize=(15, 6))  # 그래프 크기: 가로 10, 세로 ㅍㅍ
plt.plot(use_df['일자'], use_df['발전량(MWh)'], marker='o')
plt.title('년월별발전량')
plt.xlabel('일자')
plt.ylabel('발전량')
plt.grid(True)
plt.show()
```

```python
plt.bar(use_df['일자'], use_df['발전량(MWh)'], color='skyblue')
plt.title('년월별발전량')
plt.xlabel('일자')
plt.ylabel('발전량')
plt.show()
```

```python
plt.scatter(use_df['일자'], use_df['발전량(MWh)'], color='red')
plt.title('년월별발전량')
plt.xlabel('일자')
plt.ylabel('발전량')
plt.show()
```

```python
# 3D 그래프 생성
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 그룹별 선 연결
for key, group in use_df.groupby('년'):
    # 점과 선 추가
    ax.scatter(group['년'], group['월'], group['발전량(MWh)'], s=50)
    ax.plot(group['년'], group['월'], group['발전량(MWh)'])

# 축 설정
ax.set_title('연도별 월별 발전량(MWh)', fontsize=16)
ax.set_xlabel('년', fontsize=12)
ax.set_ylabel('월', fontsize=12)
ax.set_zlabel('발전량(MWh)', fontsize=12)

# 범례 표시
ax.legend()

plt.show()
```

```python
# 2차원 그래프 생성
plt.figure(figsize=(10, 6))

# 월별로 데이터 그룹화 후 그래프 그리기
for key, group in use_df.groupby('년'):
    plt.plot(group['월'], group['발전량(MWh)'], marker='o', label=f'{key}년')

# 그래프 꾸미기
plt.title('월별 발전량 (년도별 비교)', fontsize=16)
plt.xlabel('월', fontsize=12)
plt.ylabel('발전량(MWh)', fontsize=12)
plt.legend(title='년도', fontsize=7,loc='upper right')
plt.grid(True)

# 그래프 출력
plt.show()
```
