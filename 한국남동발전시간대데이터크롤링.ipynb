```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
import time
import os

DRIVER_PATH = "C:/chromedriver-win64/chromedriver.exe"

# URL 설정
URL = "https://www.koenergy.kr/kosep/gv/nf/dt/nfdt21/main.do"

options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
options.add_argument("--start-maximized")

service = Service(DRIVER_PATH)

# Chrome WebDriver 초기화
browser = webdriver.Chrome(service=service, options=options)

browser.get(URL)
```

```python
for i in range(len(start_dates)):
    # 조회 기간 입력
    start_date_input = WebDriverWait(browser, 10).until(
        EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div[3]/div[3]/section/section/div/div[2]/div[1]/form/table/tbody/tr[2]/td/div/span/input[1]"))
    )
    end_date_input = WebDriverWait(browser, 10).until(
        EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div[3]/div[3]/section/section/div/div[2]/div[1]/form/table/tbody/tr[2]/td/div/span/input[2]"))
    )
    
    browser.execute_script("arguments[0].value = arguments[1];", start_date_input, start_dates[i])
    time.sleep(2)
    browser.execute_script("arguments[0].value = arguments[1];", end_date_input, end_dates[i])
    time.sleep(2)
    
    
    
    # 검색 버튼 클릭
    search_button = WebDriverWait(browser, 10).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div[3]/div[3]/section/section/div/div[2]/div[2]/a[1]"))
    )
    search_button.click()
    time.sleep(5)
    
    
    # 엑셀 다운로드
    excel_download_button = WebDriverWait(browser, 10).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div[3]/div[3]/section/section/div/div[2]/div[3]/button[1]"))
    )
    excel_download_button.click()
    
    time.sleep(6)  # 다운로드 대기
    
    
    # CSV 다운로드
    csv_download_button = WebDriverWait(browser, 10).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div[3]/div[3]/section/section/div/div[2]/div[3]/button[2]"))
    )
    csv_download_button.click()
    
    time.sleep(6)  # 다운로드 대기
    
    
    
    # 년월일로부터 필요한 부분 추출 (예: "2410")
    formatted_date = start_dates[i][2:6]
    time.sleep(3)
    # 변경 전 파일명 및 경로
    folder_path = "C:/Users/juh08/Downloads"  # 파일이 저장된 폴더 경로
    csv_old_name = "한국남동발전_시간대별_태양광_발전실적.csv"
    xls_old_name = "한국남동발전_시간대별_태양광_발전실적.xls"
    
    # 변경 후 파일명
    csv_new_name = f"한국남동발전_시간대별_태양광_발전실적_경상대_{formatted_date}.csv"
    xls_new_name = f"한국남동발전_시간대별_태양광_발전실적_경상대_{formatted_date}.xls"
    
    # 파일 경로
    csv_old_path = os.path.join(folder_path, csv_old_name)
    xls_old_path = os.path.join(folder_path, xls_old_name)
    csv_new_path = os.path.join(folder_path, csv_new_name)
    xls_new_path = os.path.join(folder_path, xls_new_name)
    
    # 파일명 변경
    try:
        if os.path.exists(csv_old_path):
            os.rename(csv_old_path, csv_new_path)
            print(f"{csv_old_name} → {csv_new_name} 로 파일명을 변경했습니다.")
        else:
            print(f"{csv_old_name} 파일이 존재하지 않습니다.")
    
        if os.path.exists(xls_old_path):
            os.rename(xls_old_path, xls_new_path)
            print(f"{xls_old_name} → {xls_new_name} 로 파일명을 변경했습니다.")
        else:
            print(f"{xls_old_name} 파일이 존재하지 않습니다.")
    except Exception as e:
        print(f"파일명 변경 중 오류 발생: {e}")
```
